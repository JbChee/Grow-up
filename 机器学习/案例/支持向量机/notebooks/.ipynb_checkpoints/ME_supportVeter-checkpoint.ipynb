#### 如何解决这个线性不可分问题呢？咱们给它映射到高维来试试

$z=x^2+y^2$. 


```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X,y = make_blobs(n_samples=100,random_state=33,cluster_std=0.8,centers=2)
# print(X)
# print(y)
plt.scatter(X[:,0],X[:,1],c = y,s = 10,cmap='autumn')
```
